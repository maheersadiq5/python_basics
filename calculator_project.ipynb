PROJECT CALCULAOR

```python
num1 = int(__builtins__.input("enter num 1 :"))
num2 = int(__builtins__.input("enter num2:"))
operation = __builtins__.input("+,-,*,/")
if operation == "+":
    result = num1 + num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
else:
    result = num1 / num2
print(result)
```
