LIST

```python
fruit=["apple","mango","banana","pineapple"]
print(fruit)
fruit.remove("mango") # Remove by value instead of index
print(fruit)
del fruit[2] # Corrected index for "pineapple" after "mango" is removed
print(fruit)
```

```python
student = ["natiq","maheer","maha","ali"]
print("student")
student.remove("maha")
print(student)
del student[0] # Corrected index to remove 'ali' after 'maha' is removed
print(student)
```
